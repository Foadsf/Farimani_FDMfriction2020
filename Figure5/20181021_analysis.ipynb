```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# normalaised mean absolute deviation function
def nmad(data, axis=None):
    mean=np.mean(data, axis)
    return np.mean(np.absolute(data - mean), axis)/mean
```

```python
%matplotlib qt

file='20181021_analysis_nonlinear.xlsx'
xls = pd.ExcelFile(file)
confs = [con for con in xls.sheet_names]
for con in confs:
    df = pd.read_excel(file, sheet_name=con)
    plt.plot(df['speed(mm/min)'],df['kin_b'], label=con)

    
plt.legend()
plt.show()
```

```python
headers=['max_a_mean','max_b_mean','kin_a_mean','kin_b_mean','max_a_nmad','max_b_nmad','kin_a_nmad','kin_b_nmad']

df2=pd.DataFrame(columns=headers, index=confs)


for con in confs:
    df = pd.read_excel(file, sheet_name=con)
    df2['max_a_mean'][con]=df['max_a'].mean()
    df2['max_a_nmad'][con]=nmad(df['max_a'].values)
    df2['max_b_mean'][con]=df['max_b'].mean()
    df2['max_b_nmad'][con]=nmad(df['max_b'].values)
    df2['kin_a_mean'][con]=df['kin_a'].mean()
    df2['kin_a_nmad'][con]=nmad(df['kin_a'].values)
    df2['kin_b_mean'][con]=df['kin_b'].mean()
    df2['kin_b_nmad'][con]=nmad(df['kin_b'].values)
    
df2
```
